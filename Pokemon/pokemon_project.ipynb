```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

> #### dataset

```python
dados = pd.read_csv('dataset/pokemon.csv')
```

> #### Conhecendo os pokemons mais poderosos do tipo 1

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10,10))
x= dados.groupby("Type 1").Total.sum().sort_values(ascending= False)
fig= sns.barplot(x.values,x.index)
fig.set_xlabel("Overall total")
fig.set_ylabel("Type")
plt.show()
```

#### <p> Conforme o grafico acima, percebemos que de todos os pokemons listados, os mais poderosos são os do tipo água.

> #### Conhecendo os mais rápidos pokemons

```python
dados[["Name","Speed"]].sort_values(by= "Speed",ascending= False).head(10)
```

#### <p> No dataset acima, percebemos que o pokemon 'DEOXYSSPPED FORME" é o pokemon mais rapido com uma velocidade pontuada em 180 pontos.

> #### Conhecendo os pokemons com melhores ataques.

```python
dados[["Name","Attack"]].sort_values(by= "Attack",ascending= False).head(10)
```

#### <p> No dataset acima, percebemos que o pokemon 'MEWTOWMEGA MEWTOW x" é o pokemon mais forte, com um ataque de 190 pontos.

> #### Conhecendo os 10 mais fortes através de um gráfico

```python

plt.figure(figsize=(20,15))
ax= sns.violinplot(x= "Type 1",y= "Attack",data= dados)
```

#### <p> No gráfico acima, as cordas do violiono tendem mais ao centtro quando são do tipo Água. 

>#### Correlação entre ataque e defesa de todos os pokemons

```python
plt.figure(figsize=(30,30))
sns.lmplot(data= dados,x= "Attack",y= "Defense")
```

#### Na relação acima, percebemos como o ataque esta relacionado a defesa, o agrupamento ao centro da linha.

# Conclusão
